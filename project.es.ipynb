#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')
localidades = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
cinturon_sur = ds[ds['level5'].isin(localidades)].copy() #(coloco la funcion copoi para hacer una copia y eliminar la advertencia que me salta en el terminal sobre modificar el dataframe original.)

cinturon_sur['precio_normalizado'] = cinturon_sur.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min())).round(2)

fuenlabrada_index = cinturon_sur.index[cinturon_sur['level5'] == 'Fuenlabrada']
precios_normalizados_fuenlabrada = cinturon_sur.loc[fuenlabrada_index, 'precio_normalizado']

leganes_index = cinturon_sur.index[cinturon_sur['level5'] == 'Leganés']
precios_normalizados_leganes = cinturon_sur.loc[leganes_index, 'precio_normalizado']

getafe_index = cinturon_sur.index[cinturon_sur['level5'] == 'Getafe']
precios_normalizados_getafe = cinturon_sur.loc[getafe_index, 'precio_normalizado']

alcorcon_index = cinturon_sur.index[cinturon_sur['level5'] == 'Alcorcón']
precios_normalizados_alcorcon = cinturon_sur.loc[alcorcon_index, 'precio_normalizado']

plt.figure(figsize= (10, 5))
plt.hist(precios_normalizados_fuenlabrada, bins=40, alpha=0.7, label='Fuenlabrada')
plt.hist(precios_normalizados_leganes, bins=40, alpha=0.7, label='Leganés')
plt.hist(precios_normalizados_getafe, bins=40, alpha=0.7, label='Getafe')
plt.hist(precios_normalizados_alcorcon, bins=40, alpha=0.7, label='Alcorcón')

plt.title('Precios normalizados cinturón sur')
plt.legend()
plt.show()
```

En el histograma actual de precios normalizados para el cinturon sur de Madrid, podemos observar como las cuatro localidades concentran la mayor parte de las casas en precios que se mantienen entre el 0 y el 0.4 de la gráfica, dandonos a entender que se concentran más en precios bajos que altos. Siendo Leganés la localidad con menos concentración en el lado derechon y con una distribución en el gráfico mas repartida, y dando a entender que sus precios, aunque también se concentran en mayor parte en los valores bajos, también varian hacia precios mas altos con mayor frecuencia que las demás localidades.  

#### Ejercicio 19. ¿Qué puedes decir sobre el precio por metro cuadrado (precio/m2) entre los municipios de 'Getafe' y 'Alcorcón'? Debes usar el subconjunto obtenido en la pregunta 14 (★★☆)

Pista: Crea una nueva columna llamada `pps` (price per square en inglés) y luego analiza los valores

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')
localidades = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
cinturon_sur = ds[ds['level5'].isin(localidades)].copy()
cinturon_sur['pps'] = (cinturon_sur['price'] / cinturon_sur['surface']).dropna().astype(int)

medias = []
localidades_1 = ['Getafe', 'Alcorcón']

getafe_index = cinturon_sur.index[cinturon_sur['level5'] == 'Getafe']
pps_media_getafe = medias.append(cinturon_sur.loc[getafe_index, 'pps'].mean().astype(int))
pps_mediana_getafe = cinturon_sur.loc[getafe_index, 'pps'].median().astype(int)

alcorcon_index = cinturon_sur.index[cinturon_sur['level5'] == 'Alcorcón']
pps_media_alcorcon = medias.append(cinturon_sur.loc[alcorcon_index, 'pps'].mean().astype(int))
pps_mediana_alcorcon = cinturon_sur.loc[alcorcon_index, 'pps'].median().astype(int)

datos = {
    'Media': [pps_media_getafe, pps_media_alcorcon],
    'Mediana': [pps_mediana_getafe, pps_mediana_alcorcon]
}
print(f'DataFrame sobre las diferencias entre la media de precios y las medianas de los mismos entre las localidades de Getafe y Alcorcón:')
print(pd.DataFrame(datos, index= ['Getafe', 'Alcorcón']))

plt.figure(figsize = (10, 5))
plt.bar(localidades_1, medias)
plt.title('Media de precios por localidad')
plt.xlabel('Localidad')
plt.ylabel('Media')
plt.show()

print(f'Como podemos obserbar, tanto en el dataframe como en el gráfico, la localidad de Alcorcón esta ligeramente por encima de la localidad de Getafe en precios por metro cuadrado.')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('assets/real_estate.csv', sep=';')
localidades = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
cinturon_sur = ds[ds['level5'].isin(localidades)].copy()

cinturon_sur['precio_normalizado'] = cinturon_sur.groupby('level5')['price'].transform(lambda x: (x - x.min()) / (x.max() - x.min())).round(2)

fuenlabrada_index = cinturon_sur.index[cinturon_sur['level5'] == 'Fuenlabrada']
precios_normalizados_fuenlabrada = cinturon_sur.loc[fuenlabrada_index, 'precio_normalizado']

leganes_index = cinturon_sur.index[cinturon_sur['level5'] == 'Leganés']
precios_normalizados_leganes = cinturon_sur.loc[leganes_index, 'precio_normalizado']

getafe_index = cinturon_sur.index[cinturon_sur['level5'] == 'Getafe']
precios_normalizados_getafe = cinturon_sur.loc[getafe_index, 'precio_normalizado']

alcorcon_index = cinturon_sur.index[cinturon_sur['level5'] == 'Alcorcón']
precios_normalizados_alcorcon = cinturon_sur.loc[alcorcon_index, 'precio_normalizado']

plt.figure(figsize= (10, 5))

plt.subplot(2, 2, 1)
plt.scatter(fuenlabrada_index, precios_normalizados_fuenlabrada)
plt.title("Precios normalizados Fuenlabrada")
plt.xlabel('Índice')
plt.ylabel('Precio Normalizado')

plt.subplot(2, 2, 2)
plt.scatter(leganes_index, precios_normalizados_leganes)
plt.title("Precios normalizados Leganés")
plt.xlabel('Índice')
plt.ylabel('Precio Normalizado')

plt.subplot(2, 2, 3)
plt.scatter(getafe_index, precios_normalizados_getafe)
plt.title("Precios normalizados Getafe")
plt.xlabel('Índice')
plt.ylabel('Precio Normalizado')

plt.subplot(2, 2, 4)
plt.scatter(alcorcon_index, precios_normalizados_alcorcon)
plt.title("Precios normalizados Alcorcón")
plt.xlabel('Índice')
plt.ylabel('Precio Normalizado')

plt.tight_layout()
plt.show()

```

```python
## Aquí: traza la coordenadas de los estados
## PON TU CÓDIGO AQUÍ:
import pandas as pd
import matplotlib.pyplot as plt
from ipyleaflet import Map, Marker, basemaps

map = Map(center = (40.29, -3.65), zoom = 11, min_zoom = 1, max_zoom = 20,
    basemap=basemaps.OpenTopoMap)

ds = pd.read_csv('assets/real_estate.csv', sep=';')
localidades = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
cinturon_sur = ds[ds['level5'].isin(localidades)].copy()


cinturon_sur = cinturon_sur.dropna(subset=['longitude', 'latitude'])

cinturon_sur['longitude'] = cinturon_sur['longitude'].astype(str).str.replace(',', '.', regex=False).str.strip()
cinturon_sur['latitude'] = cinturon_sur['latitude'].astype(str).str.replace(',', '.', regex=False).str.strip()

cinturon_sur['longitude'] = pd.to_numeric(cinturon_sur['longitude'], errors='coerce')
cinturon_sur['latitude'] = pd.to_numeric(cinturon_sur['latitude'], errors='coerce')


diccionario_cordenadas = {}
colores = {'Fuenlabrada': 'red', 'Leganés': 'blue', 'Getafe': 'green', 'Alcorcón': 'yellow'}

fuenlabrada_index = cinturon_sur.index[cinturon_sur['level5'] == 'Fuenlabrada']
longitude_fuenlabrada = cinturon_sur.loc[fuenlabrada_index, 'longitude'].mean()
latitude_fuenlabrada = cinturon_sur.loc[fuenlabrada_index, 'latitude'].mean()
diccionario_cordenadas['Fuenlabrada'] = (latitude_fuenlabrada, longitude_fuenlabrada)

leganes_index = cinturon_sur.index[cinturon_sur['level5'] == 'Leganés']
longitude_leganes = cinturon_sur.loc[leganes_index, 'longitude'].mean()
latitude_leganes = cinturon_sur.loc[leganes_index, 'latitude'].mean()
diccionario_cordenadas['Leganés'] = (latitude_leganes, longitude_leganes)

getafe_index = cinturon_sur.index[cinturon_sur['level5'] == 'Getafe']
longitude_getafe = cinturon_sur.loc[getafe_index, 'longitude'].mean()
latitude_getafe = cinturon_sur.loc[getafe_index, 'latitude'].mean()
diccionario_cordenadas['Getafe'] = (latitude_getafe, longitude_getafe)

alcorcon_index = cinturon_sur.index[cinturon_sur['level5'] == 'Alcorcón']
longitude_alcorcon = cinturon_sur.loc[alcorcon_index, 'longitude'].mean()
latitude_alcorcon = cinturon_sur.loc[alcorcon_index, 'latitude'].mean()
diccionario_cordenadas['Alcorcón'] = (latitude_alcorcon, longitude_alcorcon)

fuenlabrada_marker = Marker(location = diccionario_cordenadas['Fuenlabrada'])
map.add_layer(fuenlabrada_marker)

leganes_marker = Marker(location=diccionario_cordenadas['Leganés'])
map.add_layer(leganes_marker)

getafe_marker = Marker(location=diccionario_cordenadas['Getafe'])
map.add_layer(getafe_marker)

alcorcon_marker = Marker(location=diccionario_cordenadas['Alcorcón'])
map.add_layer(alcorcon_marker)

map
```
